```python
# Importation des bibliothèques nécessaires
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Chargement des données
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/titanic.csv"
titanic_data = pd.read_csv(url)

# Affichage des premières lignes du dataframe
titanic_data.head()
```

# Répartition des passagers par sexe :



- Type d'analyse : Analyse univariée (qualitative)
- Intérêt : Permet de visualiser la répartition des passagers du Titanic en fonction de leur sexe. Cela peut fournir des informations sur la composition démographique du Titanic.


```python

# Visualisation des données avec Matplotlib
plt.figure(figsize=(10, 6))
sns.countplot(data=titanic_data, x='sex')
plt.title('Répartition des passagers du Titanic par sexe')
plt.xlabel('Sexe')
plt.ylabel('Nombre de passagers')
plt.show()
```

# Distribution des âges des passagers :

- Type d'analyse : Analyse univariée (quantitative)
- Intérêt : Permet de comprendre la distribution des âges des passagers du Titanic. Cela peut être utile pour identifier des tendances générales, telles que la proportion de passagers jeunes ou âgés à bord.

```python


# Visualisation des données avec Seaborn
plt.figure(figsize=(10, 6))
sns.histplot(data=titanic_data, x='age', bins=20, kde=True)
plt.title('Distribution des âges des passagers du Titanic')
plt.xlabel('Âge')
plt.ylabel('Nombre de passagers')
plt.show()


```

# Relation entre l'âge et le tarif du billet :

- Type d'analyse : Analyse bivariée (quantitative-quantitative)
- Intérêt : Explore la relation entre l'âge des passagers et le tarif de leur billet. Cela peut aider à déterminer s'il y a une corrélation entre ces deux variables, par exemple si les passagers plus âgés ont tendance à payer des tarifs de billet plus élevés.


```python


# Visualisation de la relation entre l'âge et le tarif du billet
plt.figure(figsize=(10, 6))
sns.scatterplot(data=titanic_data, x='age', y='fare', hue='survived')
plt.title('Relation entre l\'âge et le tarif du billet (survivant en couleur)')
plt.xlabel('Âge')
plt.ylabel('Tarif du billet')
plt.show()
```

# Relation entre l'âge et le sexe des passagers :

- Type d'analyse : Analyse bivariée (quantitative-qualitative)
- Intérêt : Compare la distribution des âges entre les passagers masculins et féminins. Cela peut permettre de repérer des différences dans la répartition des âges entre les sexes.

```python
# Visualisation de la relation entre l'âge (quantitative) et le sexe (qualitative)
plt.figure(figsize=(10, 6))
sns.boxplot(data=titanic_data, x='sex', y='age')
plt.title('Relation entre l\'âge et le sexe des passagers du Titanic')
plt.xlabel('Sexe')
plt.ylabel('Âge')
plt.show()
```

# Relation entre le sexe et le statut de survie des passagers :

- Type d'analyse : Analyse bivariée (qualitative-qualitative)
- Intérêt : Examine la relation entre le sexe des passagers et leur statut de survie. Cela peut aider à déterminer s'il y a eu des disparités dans les taux de survie entre hommes et femmes.


```python
# Visualisation de la relation entre le sexe (qualitative) et le statut de survie (qualitative)
plt.figure(figsize=(10, 6))
sns.countplot(data=titanic_data, x='survived', hue='sex')
plt.title('Relation entre le sexe et le statut de survie des passagers du Titanic')
plt.xlabel('Statut de survie')
plt.ylabel('Nombre de passagers')
plt.legend(title='Sexe')
plt.show()
```

# Relation entre classe, sexe et statut de survie des passagers :


- Type d'analyse : Analyse croisée (qualitative-qualitative-qualitative)
- Intérêt : Examine comment le statut de survie des passagers varie en fonction de leur classe et de leur sexe. Cela peut aider à identifier des tendances spécifiques, telles que des différences dans les taux de survie entre les classes et les sexes.

```python





# Visualisation croisée entre variables qualitatives
plt.figure(figsize=(10, 6))
sns.catplot(data=titanic_data, x='class', hue='survived', col='sex', kind='count')
plt.suptitle('Relation entre la classe, le sexe et le statut de survie des passagers du Titanic')
plt.xlabel('Classe')
plt.ylabel('Nombre de passagers')
plt.show()
```
